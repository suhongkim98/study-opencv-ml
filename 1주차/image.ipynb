# 2. Gray 영상

```python
gray = np.array([[0,128,0],
                [255,255,255],
                 [0,128,0]] )
plt.imshow(gray, cmap='gray')
```

```python
plt.imshow(gray)
```

```python
gray[0,0] = 200
plt.imshow(gray, cmap='gray')
```

```python
gray[:,:] = 128
print(gray)
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
```

# 4. 데이터 타입

```python
a = np.array([[10,20,30],
              [200,210,200]  ])
print(a.dtype)
plt.imshow(a, cmap='gray')
```

```python
a = np.array([[10,20,30],
              [200,210,200]  ], dtype='uint8')
print(a.dtype)
plt.imshow(a, cmap='gray')
```

```python
a = np.array([[10.0,20,30],
              [200,210,200]  ]) / 255
print(a.dtype)
print(a)
plt.imshow(a, cmap='gray')
```

```python
img = cv2.imread('image.jpg', 0)
print(img.shape)
plt.imshow(img, cmap='gray')
```

```python
def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()   
```

```python
imshow("", img)
```

```python
img = cv2.imread('image.jpg')
print(img.shape)

R = np.zeros(img.shape, img.dtype)
G = np.zeros(img.shape, img.dtype)
B = np.zeros(img.shape, img.dtype)

R[:,:,0] = img[:,:,2]
G[:,:,1] = img[:,:,1]
B[:,:,2] = img[:,:,0]


plt.figure(figsize=(20, 10))  #가로 세로 인치
plt.subplot(1,3,1)
plt.imshow(R)
plt.subplot(1,3,2)
plt.imshow(G)
plt.subplot(1,3,3)
plt.imshow(B)
```
