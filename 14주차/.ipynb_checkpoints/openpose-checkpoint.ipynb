# 객체검출

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
modelConfiguration = "yolov3.cfg" # yolo3 공식 홈페이지에서 다운 가능
modelWeights = "yolov3.weights"

net = cv2.dnn.readNetFromDarknet(modelConfiguration, modelWeights)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU) # cpu를 사용하겠다
```

```python
cap = cv2.VideoCapture('animals.jpg') # 숫자를 넣으면 카메라 캠으로부터 가져오기 가능
#cap = cv2.VideoCapture('cars.jpg')

hasFrame, frame = cap.read()
print(frame.shape)
```

```python
# yolo3는 416x416크기가 입력이 되어야 함, [0,0,0]은 평균, true옵션줘서 RB채널 swap해주겠다.
#crop은 이미지 크기가 맞지 않으면 자를건지 선택하는 옵션
blob = cv2.dnn.blobFromImage(frame, 1/255, (416, 416), [0,0,0], True, crop=False)  # 전처리과정
print(blob.shape)
net.setInput(blob)

# (테스트이미지 개수, 칼라, 크기, 크기)
```

```python
%%time
#yolo는 입력이 3개, 네트워크를 통과시켜 아웃풋을 받는다.
outs = net.forward(['yolo_82', 'yolo_94', 'yolo_106'])
```

```python
classesFile = "coco.names" # 80개의 클래스 이름을 기록해둔 파일
classes = None
with open(classesFile, 'rt') as f:
    classes = f.read().rstrip('\n').split('\n')
print(classes)
```

```python
def drawPred(frame, classId, conf, left, top, right, bottom):
    cv2.rectangle(frame, (left, top), (right, bottom), (255, 178, 50), 3) #박스그리기
    
    #label = '%.2f' % conf        
    label = '%s:%.2f' % (classes[classId], conf) # 클래스이름, 확률

    labelSize, baseLine = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
    top = max(top, labelSize[1])
    cv2.rectangle(frame, (left, top - round(1.5*labelSize[1])), (left + round(1.5*labelSize[0]), top + baseLine), (255, 255, 255), cv2.FILLED)
    cv2.putText(frame, label, (left, top), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0,0,0), 1)

def postprocess(frame, outs):
    confThreshold = 0.5  #Confidence threshold
    nmsThreshold = 0.4   #Non-maximum suppression threshold

    frameHeight = frame.shape[0]
    frameWidth = frame.shape[1]

    # Scan through all the bounding boxes output from the network and keep only the
    # ones with high confidence scores. Assign the box's class label as the class with the highest score.
    classIds = []
    confidences = []
    boxes = []
    for out in outs:            # 3번 loop
        for detection in out:         # 507번, 2028, 8112,, detection = 85차원
            scores = detection[5:]      #80개 클래스에 대한 확률
            classId = np.argmax(scores)   #확률이 가장큰 인덱스 가져오기
            confidence = scores[classId]   # 점수가 가장 큰 인덱스 확률값
            if confidence > confThreshold: # 확률이 50% 이상이면 검출되게끔
                center_x = int(detection[0] * frameWidth) #detection[0]은 졍규화되어있어서 frameWidth를 곱해 구한다.
                center_y = int(detection[1] * frameHeight)
                width = int(detection[2] * frameWidth)
                height = int(detection[3] * frameHeight)
                left = int(center_x - width / 2)
                top = int(center_y - height / 2)
                
                #객체가 검출됐으면 classId, 확률, 사각좌표를 각각의 리스트에 넣는다.
                classIds.append(classId)
                confidences.append(float(confidence))
                boxes.append([left, top, width, height])

    # 저렇게 박스를 그려도 개, 조금 오른쪽으로 치우쳐서 그려도 개, 이렇게 겹치는 문제를 해결해보자
    # NMSBoxes는 겹치는 박스를 그룹핑하고 확률이 가장 큰 박스 빼고 다 버리도록 하는함수
    # Perform non maximum suppression to eliminate redundant overlapping boxes with
    # lower confidences.
    # nms가 너무 낮으면  겹치는 객체를 감지하지 못함
    # nms가 너무 크면 같은 객체에 대해서 box 많이 사용
    indices = cv2.dnn.NMSBoxes(boxes, confidences, confThreshold, nmsThreshold)
    for i in indices.flatten():
        box = boxes[i]
        left = box[0]
        top = box[1]
        width = box[2]
        height = box[3]   
        drawPred(frame, classIds[i], confidences[i], left, top, left + width, top + height)
```

```python
postprocess(frame, outs)
img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# OpenPose

```python
# MPI
protoFile = "pose_deploy_linevec_faster_4_stages.prototxt"
weightsFile = "pose_iter_160000.caffemodel"
net = cv2.dnn.readNetFromCaffe(protoFile, weightsFile)
```

```python
frame = cv2.imread("yoga.jpg")
frameCopy = np.copy(frame)
frameWidth = frame.shape[1]
frameHeight = frame.shape[0]
threshold = 0.1
```

```python
inpBlob = cv2.dnn.blobFromImage(frame, 1.0 / 255, (368, 368),
                          (0, 0, 0), swapRB=False, crop=False)

net.setInput(inpBlob)
```

```python
%%time
# yolo랑은 다르게 출력층이 1개라서 파라미터가 없음
output = net.forward()
```

```python
print(output.shape)
H = output.shape[2]
W = output.shape[3]

#(테스트이미지 개수, 44개의 확률맵, 각각의확률맵크기는 46, 46)
```

```python
print(output[0, 0, :, :])
```

```python
plt.imshow(output[0, 0, :, :]) # 머리 꼭대기 부분의 확률값, 각각의 포인트는 0~14까지있음
```

```python
plt.imshow(output[0, 15, :, :]) # 0부터 14번까지 누적시킨 것 출력
```

```python
# Empty list to store the detected keypoints

nPoints = 15

points = []
for i in range(nPoints):
    # confidence map of corresponding body's part.
    probMap = output[0, i, :, :]

    # Find global maxima of the probMap.
    minVal, prob, minLoc, point = cv2.minMaxLoc(probMap)
    
    # Scale the point to fit on the original image     
    # 원래 이미지 크기의 좌표 구하기
    x = frameWidth * ((point[0]) / W) # (point[0]) / W로 정규화해서 frameWidth를 곱해 원래 x를 구한다
    y = frameHeight * ((point[1]) / H)

    if prob > threshold : # 확률이 0.1보다 크면 검출되게끔
        cv2.circle(frameCopy, (int(x), int(y)), 8, (0, 255, 255), thickness=-1, lineType=cv2.FILLED)
        cv2.putText(frameCopy, "{}".format(i), (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, lineType=cv2.LINE_AA)
        cv2.circle(frame, (int(x), int(y)), 8, (0, 0, 255), thickness=-1, lineType=cv2.FILLED)

        # Add the point to the list if the probability is greater than the threshold
        points.append((int(x), int(y)))
    else :
        points.append(None)


plt.figure(figsize=[10,10])
plt.imshow(cv2.cvtColor(frameCopy, cv2.COLOR_BGR2RGB))
```

```python
# 0,1 연결, 1,2 연결 2,3 연결.............. 정해주어야 한다.
POSE_PAIRS = [[0,1], [1,2], [2,3], [3,4], [1,5], [5,6], [6,7], [1,14], [14,8], [8,9], [9,10], [14,11], [11,12], [12,13] ]

for pair in POSE_PAIRS:
    partA = pair[0]
    partB = pair[1]

    if points[partA] and points[partB]:
        cv2.line(frame, points[partA], points[partB], (0, 255, 255), 3) # 이걸통해 그리자

plt.figure(figsize=[10,10])
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```
